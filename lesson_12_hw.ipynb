```python
df['age'] = np.random.randint(18, 66, size=len(df))
df
```

```python
age_bins = [18, 25, 35, 50, 65]
age_labels = ['18-25', '26-35', '36-50', '51+']
df['agegroup'] = pd.cut(df['age'], bins=age_bins, labels=age_labels, right=True)
df
```

```python
df.groupby('agegroup', observed=True).size()
```

```python
df.groupby('agegroup', observed=True)['age'].mean()
```
