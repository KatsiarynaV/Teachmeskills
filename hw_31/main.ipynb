```python
from confluent_kafka.admin import AdminClient, NewTopic

conf = {
    'bootstrap.servers': 'localhost:9094'
}

admin = AdminClient(conf)
```

```python
new_topic = NewTopic(
    topic='raw_data', 
    num_partitions=3,
    replication_factor=1
)

topic_temperature = NewTopic(
    topic='topic_temperature', 
    num_partitions=3,
    replication_factor=1
)

topic_humidity = NewTopic(
    topic='topic_humidity', 
    num_partitions=3,
    replication_factor=1
)

fs = admin.create_topics([new_topic, topic_temperature, topic_humidity])

for topic, f in fs.items():
    try:
        f.result()
        print(f"[v] ТTopics {topic} created")
    except:
        print("Error")
```

```python
import json
import random
import time
from datetime import datetime
from confluent_kafka import Producer

conf = {
    'bootstrap.servers': 'localhost:9094'
}

producer = Producer(conf)

TOPIC = 'raw_data'


def generate_message():
    msg_type = random.choice(["temperature", "humidity"])
    value = round(random.uniform(10.0, 30.0) if msg_type == "temperature" else round(random.uniform(30.0, 80.0), 1))
    return {
        "type": msg_type,
        "value": value,
        "timestamp": datetime.now().isoformat()
    }

try:
    print("[v] Producer start working..")
    while True:
        msg = generate_message()
        producer.produce(TOPIC, value=json.dumps(msg))
        producer.poll(0)
        time.sleep(0.1)

        print("[v] 1 message sent", datetime.now().isoformat())

except KeyboardInterrupt:
    print("[x] Producer stopped")
finally:
    producer.flush()
```
